```python
import scipy 
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
final_image = "0002"

image_path = f'ISRwall/input/img_{final_image}.jpg'  # Replace with your image path
image = cv2.imread(image_path)

# Convert the image to RGB (OpenCV reads images in BGR format)
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# Load the image

data_input_folder = 'ISRwall/input/'
data_input_01 = scipy.io.loadmat(data_input_folder + f'kp_{final_image}.mat')

data_input_01_desc = data_input_01['desc']
data_input_01_kp = data_input_01['kp']


# Step 3: Draw circles on the image at each key point
for point in data_input_01_kp:
    x, y = int(point[0]), int(point[1])
    radius = 10             # Radius of the circle
    color = (255, 0, 0)     # Red color in RGB
    thickness = 3           # Thickness of the circle border
    cv2.circle(image_rgb, (x, y), radius, color, thickness)

# Step 4: Display the result using Matplotlib
plt.figure(figsize=(10, 6))
plt.imshow(image_rgb)
plt.axis('off')  # Hide axis
plt.title('Image with Key Points')
plt.show()
```
