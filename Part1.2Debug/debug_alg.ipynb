```python
import heapq

graph = {
    'A': {'B': 1, 'C': 4},
    'B': {'A': 1, 'D': 2, 'E': 5},
    'C': {'A': 4, 'F': 3},
    'D': {'B': 2},
    'E': {'B': 5, 'F': 1},
    'F': {'C': 3, 'E': 1}
}

nodes_data = {
    'A': (0, 0),
    'B': (1, 1),
    'C': (0, 4),
    'D': (2, 1),
    'E': (1, 5),
    'F': (0, 7)
}

def heuristic(node1, node2, nodes_data):
    x1, y1 = nodes_data[node1]
    x2, y2 = nodes_data[node2]
    return abs(x1 - x2) + abs(y1 - y2)  # Manhattan distance

def k_beam_search(graph, start_node, end_node, nodes_data, k):
    # Priority queue to store (heuristic cost, current_node, path)
    beam = [(0, start_node, [start_node])]
    visited = set()

    while beam:
        new_beam = []
        for _, current_node, path in beam:
            if current_node in visited:
                continue

            visited.add(current_node)

            if current_node == end_node:
                return path

            for neighbor, cost in graph.get(current_node, {}).items():
                if neighbor not in visited:
                    estimated_cost = heuristic(current_node, neighbor, nodes_data)
                    heapq.heappush(new_beam, (estimated_cost, neighbor, path + [neighbor]))

        # Keep only the top K nodes in the beam
        beam = heapq.nsmallest(k, new_beam)

    return None

# Example usage:
start_node = 'A'
end_node = 'F'
k = 5  # Beam width
path = k_beam_search(graph, start_node, end_node, nodes_data, k)
print("Path:", path)
```
